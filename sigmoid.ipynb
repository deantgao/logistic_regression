```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sigmoid = lambda z: 1/(1+np.exp(-z))
sig_a = lambda z: (np.exp(z) - np.exp(-z))/(np.exp(z)+np.exp(-z))
z = [x for x in range(-10,11)]
fz = []
sig = []
for i in z:
    fz.append(sig_a(i))
    sig.append(sigmoid(i))
```

```python
plt.plot(z, fz)
plt.plot(z, sig)
plt.xlabel('z')
plt.ylabel('f(z)')
plt.show()
```
